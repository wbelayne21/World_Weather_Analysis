```python
import random
import numpy as np
from citipy import citipy
from config import weather_api_key
from config2 import weather_apiKey
import requests
import pandas as pd
```

```python
lats = np.random.uniform(high = 90, low = -90, size = 2000)
lngs = np.random.uniform(high = 180, low = -180, size = 2000)
```

```python
coordinates = zip(lats, lngs)
```

```python
cities = []
```

```python
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    if city not in cities:
        cities.append(city)
```

```python
len(cities)
```

```python
#base_url = "http://api.openweathermap.org/data/2.5/json" 
#base_url = "http://api.openweathermap.org/data/2.5/weather?"
base_url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_apiKey
#print(base_url)

params = {
   "location": coordinates,
    "key": weather_apiKey
}
city_url = requests.get(base_url, params = params).json()
city_url

######### this is not running. I genereated a new API and tried to use that but error in first cell "cannot import"
######### created a second textfile and saved the new api_key there but now says invalid API
```

```python
city_data = []
# Print to logger
print("Beginning Data Retrieval     ")
print("-----------------------------")
# Create counters
record_count = 1
set_count = 1
# Loop through all the cities in our list
for i, city in enumerate(cities):
   # Group cities in sets of 50 for logging purposes
   if (i % 50 == 0 and i >= 50):
       set_count += 1
       record_count = 1
   # Create endpoint URL with each city
   city_url = base_url + "&q=" + city.replace(" ","+")
   # Log the url, record, and set numbers
   print(f"Processing Record {record_count} of Set {set_count} | {city}")
   # Add 1 to the record count
   record_count += 1
   # Run an API request for each of the cities
   try:
       city_weather = requests.get(city_url).json()
       # Parse out the max temp, humidity, and cloudiness
       city_lat = city_weather["coord"]["lat"]
       city_lng = city_weather["coord"]["lon"]
       city_max_temp = city_weather["main"]["temp_max"]
       city_humidity = city_weather["main"]["humidity"]
       city_clouds = city_weather["clouds"]["all"]
       city_wind = city_weather["wind"]["speed"]
       city_country = city_weather["sys"]["country"]
       city_weather_description = city_weather["weather"][0]["description"]
           # Append the City information into city_data list
       city_data.append({"City": city.title(),
                         "Lat": city_lat,
                         "Lng": city_lng,
                         "Max Temp": city_max_temp,
                         "Current Description": city_weather_description,
                         "Humidity": city_humidity,
                         "Cloudiness": city_clouds,
                         "Wind Speed": city_wind,
                         "Country": city_country})
   # If an error is experienced, skip the city
   except:
       print("City not found. Skipping...")
       pass
# Indicate that Data Loading is complete
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
# adding data into a new df
city_data_df = pd.DataFrame(city_data)
city_data_df.head()
```

```python
# reorder columns
new_column_order = ["City", "Country", "Lat","Lng", "Max Temp", "Humidity", "Cloudiness", "Wind Speed", "Current Description"]
```

```python
city_data_df = city_data_df[new_column_order]
city_data_df.head(10)
```

```python
#Exporting the data 
output_data_file = "Weather_Database_csv/WeatherPy_Database.csv"
city_data_df.to_csv(output_data_file, index_label="City_ID")
```
